```python
import pandas as pd
import numpy as np
import sklearn

dataset = pd.read_csv("新竹_2019.csv",encoding='big5')
dataset = dataset.drop([0])
dataset = dataset.drop(columns="測站")

d10 = dataset[dataset["日期"].str.contains('/10/')]
d11 = dataset[dataset["日期"].str.contains('/11/')]
d12 = dataset[dataset["日期"].str.contains('/12/')]

# dataset.count()
```

```python
d1011 = d10.append(d11)
# d1011
d12
```

```python
# 10 11 month
train = d1011.iloc[0:18]
train = train.drop(columns = ["測項","日期"])
d1011 = d1011.drop(train.index, axis=0)
d1011 = d1011.drop(columns = ["測項","日期"])
# d1011
for i in range(60):
    d = d1011.iloc[0:18]
    train = pd.concat([train, d.set_index(train.index)], axis=1)  #merge two df while using dd's index to concat
    d1011 = d1011.drop(d.index, axis=0)
    
train
# d1011
```

```python
test = d12.iloc[0:18]
d12 = d12.drop(columns = ["測項","日期"])
test = test.drop(columns = ["測項","日期"])
for i in range(31):
    t = d12.iloc[0:18]
    test = pd.concat([test, t.set_index(test.index)], axis=1) 
    d12 = d12.drop(t.index, axis=0)
test
# d12
```

```python
#---data cleaning---
# below method is using .sub ways to substitute the abnormal token,
# but in this assignment we want to replace the abnormal value to NaN
# import re
# def sub(a):               #a is a string 
#     a = re.sub(r"[^0-9,.]", "", a)       #remove every character but 0-9 and .
#                                         #good regex website for testing : https://regex101.com/r/LcDOH9/1
#     return a
```

```python
import re

def sub(value):
    a = re.search(r"A", value) 
    b = re.search(r"#", value)       
    c = re.search(r"x", value)
    d = re.search(r"\*", value)
    if(a or b or c or d):
        value = np.nan          #we need to convert abnormal value with dataframe's version of none (np.nan)
    return value
```

```python
for index, row in train.iterrows():        #row by row to iterate
    for i in range(0,len(row)):            #through each row
        row[i] = sub(str(row[i]))
# train
for index, row in test.iterrows():        
    for i in range(0,len(row)):            
        row[i] = sub(str(row[i]))
test
#---data cleaning end---
```

```python
# fill the NaN with average of previous and next rows by using interpolate method

train = train.astype(np.float)
train = train.interpolate(method='linear', axis=1)     #In interpolate axis=1 means row by row

test = test.astype(np.float)
test = test.interpolate(method='linear', axis=1)
test
# train
```

```python
from sklearn.metrics import mean_absolute_error     #MAE
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
test_pm = test.iloc[9]
t_p_1 = []
true_p_1 = []
for i in range(len(test.columns)-6):
    t_p_1.append(test_pm.iloc[i: i+6].to_numpy())
    true_p_1.append(test_pm.iloc[i+6])
    
t_p_6 = []
true_p_6 = [] 
for i in range(len(test.columns)-11):
    t_p_6.append(test_pm.iloc[i: i+6].to_numpy())
    true_p_6.append(test_pm.iloc[i+11])
        
    
    
t_all_1 = []
true_all_1 = []
for i in range(len(test.columns)-6):
    tmp = []
    tmp_y = []
    for j in range(18):
        tmp_y.append(test.iloc[j,i+6])
        for k in range(6):
            tmp.append(test.iloc[j,i+k])
    t_all_1.append(tmp)
    true_all_1.append(tmp_y)    
    

t_all_6 = []
true_all_6 = []
for i in range(len(test.columns)-11):
    tmp = []
    tmp_y = []
    for j in range(18):
        tmp_y.append(test.iloc[j,i+11])
        for k in range(6):
            tmp.append(test.iloc[j,i+k])
    t_all_6.append(tmp)
    true_all_6.append(tmp_y)  

#----1--pm 2.5-----
lm = LinearRegression()
lm = lm.fit(x_p_1, y_p_1)
pm_predicted = lm.predict(t_p_1)
print("1 LinearRegression for pm 2.5", mean_absolute_error(true_p_1, pm_predicted))

regr = RandomForestRegressor(max_depth=2, random_state=0)
regr.fit(x_p_1,y_p_1)
pm_predicted = regr.predict(t_p_1)
print("1 RandomForestRegressor for pm 2.5:", mean_absolute_error(true_p_1, pm_predicted))

#----1--pm 2.5-----

#----1--all 18 rows--
lm = lm.fit(x_all_1, y_all_1)
pm_predicted = lm.predict(t_all_1)
print("1 LinearRegression for all 18 rows:", mean_absolute_error( true_all_1, pm_predicted))

regr = RandomForestRegressor(max_depth=2, random_state=0)
regr.fit(x_all_1,y_all_1)
pm_predicted = regr.predict(t_all_1)
print("1 RandomForestRegressor for all 18 rows:", mean_absolute_error(true_all_1, pm_predicted))


#----1--all 18 rows--

#----6--pm 2.5-----
lm = LinearRegression()
lm = lm.fit(x_p_6, y_p_6)
pm_predicted = lm.predict(t_p_6)
print("6 LinearRegression for pm 2.5", mean_absolute_error(true_p_6, pm_predicted))

regr = RandomForestRegressor(max_depth=2, random_state=0)
regr.fit(x_p_6,y_p_6)
pm_predicted = regr.predict(t_p_6)
print("6 RandomForestRegressor for pm 2.5:", mean_absolute_error(true_p_6, pm_predicted))

#----6--pm 2.5-----

#----6--all 18 rows--
lm = lm.fit(x_all_6, y_all_6)
pm_predicted = lm.predict(t_all_6)
print("6 LinearRegression for all 18 rows:", mean_absolute_error( true_all_6, pm_predicted))

regr = RandomForestRegressor(max_depth=2, random_state=0)
regr.fit(x_all_6,y_all_6)
pm_predicted = regr.predict(t_all_6)
print("6 RandomForestRegressor for all 18 rows:", mean_absolute_error(true_all_6, pm_predicted))


#----6--all 18 rows--
```
